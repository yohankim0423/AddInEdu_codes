```python
def show_histogram(img, title, xmax, ymax=None):
    color = ('b', 'g', 'r')
    for i, clr in enumerate(color):
        histr = cv2.calcHist([img], channels=[i], mask=None, histSize=[256], ranges=[0, 256])
        plt.plot(histr, color = clr)
        plt.xlim([0, xmax])
        if ymax != None:
            plt.ylim([0, ymax])
    plt.title(title)
```

```python
show_histogram(bricks, 'HISTOGRAM FOR BLUE BRICKS', 256)
```

```python
show_histogram(dark_horse, 'HISTOGRAM FOR DARK HORSE', 256)
```

```python
show_histogram(dark_horse, 'HISTOGRAM FOR DARK HORSE', 15, 50000)
```

2번 히스토그램 평활화 equalization

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_img(img, cmap=None):
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
gorilla = cv2.imread('/home/verlo/dev_ws/OpenCV/data/gorilla.jpg', cv2.IMREAD_GRAYSCALE)
display_img(gorilla, cmap='gray')
```

```python
hist_values = cv2.calcHist([gorilla], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
plt.plot(hist_values);
```

```python
eq_gorilla = cv2.equalizeHist(gorilla)
display_img(eq_gorilla, cmap='gray')
```

```python
hist_values = cv2.calcHist([eq_gorilla], channels=[0], mask=None, histSize=[256], ranges=[0,256])
plt.plot(hist_values)
```

```python
colour_gorilla = cv2.imread('/home/verlo/dev_ws/OpenCV/data/gorilla.jpg')
show_colour_gorilla = cv2.cvtColor(colour_gorilla, cv2.COLOR_BGR2RGB)
plt.imshow(show_colour_gorilla);
```

```python
hsv = cv2.cvtColor(colour_gorilla, cv2.COLOR_BGR2HSV)
hsv[:,:,2] = cv2.equalizeHist(hsv[:,:,2])
equalized_colour_gorilla = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
display_img(equalized_colour_gorilla)
```
