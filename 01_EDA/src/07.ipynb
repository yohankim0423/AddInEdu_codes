```python
import pandas as pd

col_names = ['스트레스', '스트레스남학생', '스트레스여학생', '우울감경험률', '우울남학생', '우울여학생', '자살생각율', '자살남학생', '자살여학생']
raw_data = pd.read_excel(url, header=1, usecols="C:K", names=col_names)
raw_data
```
