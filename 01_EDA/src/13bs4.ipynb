```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = "https://ko.wikipedia.org/wiki/%EA%B5%AC%EB%AF%B8%ED%98%B8%EB%8E%90_1938"
page = urlopen(url)

soup = BeautifulSoup(page, "html.parser")
print(soup.prettify())
```

```python
import urllib
from urllib.request import Request

url = "https://ko.wikipedia.org/wiki/{search_words}"
req = Request(url.format(search_words = urllib.parse.quote("구미호뎐 1938")))
res = urlopen(req)

soup = BeautifulSoup(res, "html.parser")
print(soup.prettify())
```

```python
for idx, each in enumerate(soup.find_all("ul")):
    print("[" + str(idx) + "]" + "==============================")
    print(each.prettify())
```
