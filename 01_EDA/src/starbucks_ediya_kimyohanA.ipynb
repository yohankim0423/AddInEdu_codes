```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
starbucks = soup.find_all(class_="quickResultLstCon")
starbucks_name = [Name.attrs["data-name"] for Name in starbucks]
starbucks_name
```

```python
starbucks = soup.find_all(class_="quickResultLstCon")
starbucks_lat = [Lat.attrs["data-lat"] for Lat in starbucks]
starbucks_lat
```

```python
starbucks = soup.find_all(class_="quickResultLstCon")
starbucks_long = [Long.attrs["data-long"] for Long in starbucks]
starbucks_long
```

```python
import re

starbucks = soup.find_all(class_="quickResultLstCon")
starbucks_address = [Address.find(class_="result_details").text for Address in starbucks]
# 전화번호 혹은 숫자와 하이픈을 포함하는 부분을 제거
result_addresses = [re.sub(r'[\d-]+$', '', address).strip() for address in starbucks_address]
# 결과 출력
for address in result_addresses:
    print(address)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import re


# 주소 클릭
time.sleep(0.5)
location_search = driver.find_element(By.CSS_SELECTOR, "#contentWrap > div.contents > div > div.store_search_pop > ul > li:nth-child(2) > a")
location_search.click()

# 검색하고자 하는 키워드 리스트
keywords = ['서울 강남구', '서울 강동구', '서울 강북구', '서울 강서구', '서울 관악구',
            '서울 광진구', '서울 구로구', '서울 금천구', '서울 노원구', '서울 도봉구',
            '서울 동대문구','서울 동작구', '서울 마포구', '서울 서대문구','서울 서초구',
            '서울 성동구', '서울 성북구', '서울 송파구', '서울 양천구', '서울 영등포구',
            '서울 용산구', '서울 은평구', '서울 종로구','서울 중구', '서울 중랑구']

# 매장명과 매장 주소를 저장할 리스트 생성
ediya_store_name = []
ediya_store_address = []

for keyword in keywords:
    # 검색창 찾기
    location_search = driver.find_element(By.CSS_SELECTOR, "#keyword")
    
    # 이전 검색어 지우기
    location_search.clear()
    
    # 검색창에 키워드 입력
    location_search.send_keys(keyword)
    
    # 검색 실행
    location_search.send_keys(Keys.ENTER)

    # 매장명과 매장 주소 가져오기
    ediya_store_names = driver.find_elements(By.CSS_SELECTOR, "#placesList").text
    ediya_store_addresses = driver.find_elements(By.CSS_SELECTOR, "#placesList").text

    # 그 태그에서 이름이랑 주소 가져오기
    for name in ediya_store_name:
        ediya_store_name.append(name.text)
        
    for address in ediya_store_address:
        cleaned_text = re.sub(r'\([^)]*\)', '', address.text)
        ediya_store_address.append(cleaned_text.text)
        # 페이지 로딩 대기
    time.sleep(5)

    # 결과 출력
for name in ediya_store_name:
    print(name)
for address in ediya_store_address:
    print(address)
```
